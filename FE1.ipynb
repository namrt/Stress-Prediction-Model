```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Preprocessed.csv')
```

```python
df.head()
```

```python
#renaming the stress score column
df.rename(columns = {'Stress Score':'Stress_Score'}, inplace = True) 
#grouping data on the basis of 'sex','Predicted_age','Weekday','TimeOfDay' to find average score for each group
grouped = df.groupby(['sex','Predicted_age','Weekday','TimeOfDay']).Stress_Score.agg(['mean'])
```
