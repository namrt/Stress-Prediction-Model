```python
#Preocess updatedAt

new = df["UserActivity_updatedAt"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["Update_date"]= new[0] 
  
# making separate last name column from new data frame 
df["Update_time"]= new[1] 
```

```python
df.head()
```

```python
time = df["Update_time"].str.split(":", n = 1, expand = True) 
df["Update_hour"]= time[0] 
df['Update_hour'] = df['Update_hour'].astype(int)
```

```python
for i in range (df.shape[0]):
    x=df.iloc[i]['Update_hour']
    if (x > 4) and (x <= 12 ):
        df.loc[i,'TimeOfDay'] ='Morning'
    elif (x > 12) and (x <= 16):
        df.loc[i,'TimeOfDay']='Noon'
    elif (x > 16) and (x <= 20) :
        df.loc[i,'TimeOfDay'] ='Eve'
    else:
        df.loc[i,'TimeOfDay']='Night'
    
```

```python
from datetime import datetime, timedelta
import calendar
df['Datetime'] = pd.to_datetime(df['UserActivity_updatedAt'])

for i in range (df.shape[0]):
    x=df.iloc[i]['Datetime']
    df.loc[i,'Weekday'] =calendar.day_name[x.weekday()]
    
```
